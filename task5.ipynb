```python
from transformers import pipeline

classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
labels = [
    "Billing",
    "Technical Issue",
    "Account Access",
    "Refund",
    "Feature Request",
    "Complaint",
    "Shipping",
    "Product Inquiry"
]
```

```python
ticket = "I paid for premium subscription but still cannot access premium features."

result = classifier(
    ticket,
    candidate_labels=labels,
    multi_label=True
)

top3 = list(zip(result["labels"][:3], result["scores"][:3]))

for tag, score in top3:
    print(tag, round(score,4))
```

```python
import pandas as pd

df = pd.read_csv("customer_support_tickets.csv")

print(df.columns)
print("\n")
print(df.head())
```

```python
def get_top3_tags(text):
    result = classifier(
        text,
        candidate_labels=labels,
        multi_label=True
    )

    return list(zip(result["labels"][:3], result["scores"][:3]))
```

```python
df_small = df.head(100).copy()
```

```python
print(df_small.shape)
print(df_small.head())
```

```python
results = []

for i, text in enumerate(df_small["Ticket Description"]):
    print(f"Processing {i+1}/{len(df_small)}")

    result = classifier(
        text,
        candidate_labels=labels,
        multi_label=True
    )

    results.append(list(zip(result["labels"][:3], result["scores"][:3])))

df_small["Top3_Predicted_Tags"] = results
```

```python
df_small.to_csv("ticket_tagged_sample_100.csv", index=False)
print("Done!")
```

```python
def format_tags(tag_list):
    return " | ".join([f"{t[0]} ({round(t[1],2)})" for t in tag_list])

df_small["Formatted_Tags"] = df_small["Top3_Predicted_Tags"].apply(format_tags)

df_small[["Ticket Description", "Formatted_Tags"]].head()
```

```python
def few_shot_predict(text):
    prompt = f"""
You are a support ticket classifier.

Examples:

Ticket: I was charged twice for my order
Tags: Billing, Refund, Complaint

Ticket: I cannot log into my account
Tags: Account Access, Technical Issue

Ticket: My product has not been delivered
Tags: Shipping, Complaint

Now classify:

Ticket: {text}
Tags:
"""

    result = classifier(
        prompt,
        candidate_labels=labels,
        multi_label=True
    )

    return list(zip(result["labels"][:3], result["scores"][:3]))
```

```python
few_results = []

for text in df_small["Ticket Description"]:
    res = few_shot_predict(text)
    few_results.append(res)

df_small["Few_Shot_Tags"] = few_results
```
